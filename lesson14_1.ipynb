```python
import matplotlib.pyplot as plt

#1.製作figure
fig = plt.figure()

#2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

#3.將圖的數據儲存在axes中
X = [0,1,2]
Y = [3,4,5]
ax_1.plot(X,Y)

#4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
ax1=figure.add_subplot(1,3,1)
ax1.set_title("label1")
ax2=figure.add_subplot(1,3,2)
ax2.set_title("label2")
ax3=figure.add_subplot(1,3,3)
ax3.set_title("label2")
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot([-1,-4.3,15,21,31])
print(axes.get_xticks())

axes.set_xticklabels(['a','b','c','d','e','f','g','h','i','j'])
plt.show()
```

```python
import matplotlib.pyplot as plt
student1 = {'ch':78,'eng':80,'m':92,'his':75,'dis':85}
subjects=list(student1.keys())
s1_value=list(student1.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,s1_value)
print(axes.get_xticks())
print(axes.get_xticklabels())
plt.show()
```
